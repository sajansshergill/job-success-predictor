```python
# 1. Load Accuracy Scores (simplified parse)
with open("/Users/sajanshergill/Desktop/job-success-predictor/outputs/model_evaluation.txt") as f:
    lines = f.readlines()

results = {}
model = None
for line in lines:
    if "Model:" in line:
        model = line.strip().split(":")[1].strip()
    if "Accuracy" in line:
        acc = float(line.strip().split(":")[1])
        results[model] = acc

# 2. Plot
import matplotlib.pyplot as plt

plt.figure(figsize=(8,4))
plt.bar(results.keys(), results.values(), color='skyblue')
plt.ylabel("Accuracy")
plt.title("Model Accuracy Comparison")
plt.xticks(rotation=15)
plt.ylim(0, 1)
```
